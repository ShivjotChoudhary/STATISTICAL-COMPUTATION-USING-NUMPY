```python
#importing packages numpy
import numpy as np
```

```python
#loading iris dataset
from sklearn.datasets import load_iris
```

```python
#loading data and features into variables
iris=load_iris()
x=iris.data
features_name= iris.feature_names
```

```python
#calculated mean
mean_value= np.mean(x)
print("This is the airthmatical mean : ",mean_value)

mean_value_contains_nan= np.nanmean(x)    #this function skips nan values and compute mean for the remaining values
print("This is the airthmatical mean with nan : ",mean_value_contains_nan)
```

**Mean analysis**

to find central tendency

```python
#calculated median
median_value= np.median(x)
print("This is the median : ",median_value)

median_value_contains_nan= np.nanmedian(x)    #this function skips nan values and compute median for the remaining values
print("This is the median with nan : ",median_value_contains_nan)
```

**MEDIAN analysis**

median is less affected from outliers  
mean ~ median = data is not heavly skewed

```python
#calculated variance
variance= np.var(x)
print("This is the variance : ",variance)

nan_variance= np.nanvar(x)    #this function skips nan values and compute variance for the remaining values
print("This is the variance with nan : ",nan_variance)
```

**Variance Analysis**  
* It shows how spread out the data is around its mean
* high variance value signifies widely spread data  
* low variance value signifies tighly packed values
* if the value is zero then the data points are identical

 > **Conclusion** : High varience means it varies the most among the flowers

```python
#calculated standard_deviation
standard_deviation= np.std(x)
print("This is the standard_deviation : ",standard_deviation)

nan_standard_deviation= np.nanstd(x)    #this function skips nan values and compute standard_deviation for the remaining values
print("This is the standard_deviation with nan : ",nan_standard_deviation)
```

**Standard deviation Analysis**

* high variance value signifies widely spread data  
* low variance value signifies tighly packed values

 more std features has more variability

##**Class-wise Statistical Analysis (Flower Type Wise)**

The Iris dataset has 3 flower species:  
Setosa (0)  
Versicolor (1)  
Virginica (2)  
We now perform statistical operations separately for each flower type and compare them

```python
y=iris.target
```

```python
setosa = x[y == 0]
versicolor = x[y == 1]
virginica = x[y == 2]
```

Splitting the dataset allows us to understand how each species differs statistically, which is crucial for classification problems.

```python
mean_setosa = np.mean(setosa, axis=0)
mean_versicolor = np.mean(versicolor, axis=0)
mean_virginica = np.mean(virginica, axis=0)
print(mean_setosa)
print(mean_versicolor)
print(mean_virginica)
```

**Mean Comparison analysis**
Setosa has smallest petal length & width  
Virginica has largest petal features  
Versicolor lies in between

```python
std_setosa = np.std(setosa, axis=0)
std_versicolor = np.std(versicolor, axis=0)
std_virginica = np.std(virginica, axis=0)
print(std_setosa)
print(std_versicolor)
print(std_virginica)

```

**Standard Deviation Comparison analysis**  
Setosa shows lowest variability → more compact cluster  
Virginica shows higher variability → more spread out  
Lower std means easier separation by ML models.

```python
var_setosa = np.var(setosa, axis=0)
var_versicolor = np.var(versicolor, axis=0)
var_virginica = np.var(virginica, axis=0)
print(var_setosa)
print(var_versicolor)
print(var_virginica)

```

**Variance Comparison analysis**  
Petal length variance increases from Setosa → Versicolor → Virginica, showing gradual complexity.

```python
range_setosa = np.ptp(setosa, axis=0)
range_versicolor = np.ptp(versicolor, axis=0)
range_virginica = np.ptp(virginica, axis=0)
print(range_setosa)
print(range_versicolor)
print(range_virginica)
```

**Range Comparison (Max − Min)  analysis**  
Virginica has the widest range, meaning its measurements vary more widely.

```python
corr_setosa = np.corrcoef(setosa.T)
corr_versicolor = np.corrcoef(versicolor.T)
corr_virginica = np.corrcoef(virginica.T)
print(corr_setosa)
print(corr_versicolor)
print(corr_virginica)

```

**Correlation Analysis per Species**  
 Strong petal length–width correlation in all species  
Correlation strength increases from Setosa to Virginica

```python
Z_setosa = (setosa - mean_setosa) / std_setosa
Z_versicolor = (versicolor - mean_versicolor) / std_versicolor
Z_virginica = (virginica - mean_virginica) / std_virginica
print(Z_setosa)
print(Z_versicolor)
print(Z_virginica)
```

**Z-Score Normalization per Species analysis**  
Each species becomes centered at 0 with unit variance, allowing fair comparison.
